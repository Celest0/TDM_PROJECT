## Question 3

```r
# 3.1 Count of 'Comedy' rows

# Exclusively 'Comedy'
exclusive_comedy <- sum(myDF$genre == "Comedy")

# Multiple genres including 'Comedy' (but not exclusively)
multi_comedy <- sum(grepl("Comedy", myDF$genre) & myDF$genre != "Comedy")

# All rows that contain 'Comedy'
all_comedy <- sum(grepl("Comedy", myDF$genre))

# Check: exclusive + multiple = total
exclusive_comedy + multi_comedy == all_comedy


# 3.2 Most frequent multi-genre 'Comedy' combination (excluding exclusive 'Comedy')
multi_comedy_table <- sort(table(myDF$genre[grepl("Comedy", myDF$genre) & myDF$genre != "Comedy"]), decreasing = TRUE)
head(multi_comedy_table, 1)   # shows the most frequent combination


# 3.3 Repeat the same for 'Romance'

# Exclusively 'Romance'
exclusive_romance <- sum(myDF$genre == "Romance")

# Multiple genres including 'Romance'
multi_romance <- sum(grepl("Romance", myDF$genre) & myDF$genre != "Romance")

# All rows that contain 'Romance'
all_romance <- sum(grepl("Romance", myDF$genre))

# Check: exclusive + multiple = total
exclusive_romance + multi_romance == all_romance
```
