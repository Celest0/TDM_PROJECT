## Question 2

```r
library(dplyr)

# Group by genre, calculate average critic score, and sort by score
avg_scores_dplyr <- video_games %>%
  group_by(genre) %>%
  summarize(avg_critic_score = mean(critic_score, na.rm = TRUE)) %>%
  arrange(desc(avg_critic_score))

# Display the table
avg_scores_dplyr
```

```r
# I prefer dplyr because its syntax is clearer and easier to read,
# especially when performing multiple steps like grouping, summarizing, and sorting.
# It also returns a tidy dataframe, which makes further analysis and visualization simpler.
# Overall, it feels more intuitive and organized than using tapply().
```
